```python
%matplotlib qt5
import hyperspy.api as hs
```

```python
s = hs.load("002_EELS-SI_1.7nm_2mmAp_MnFe_Stack_Average denoised.dm3")
s
```

```python
s.plot()
```

```python
s = s.inav[1:-1, 1:-1]
s
```

```python
s.plot()
```

```python
s.set_microscope_parameters(beam_energy=300, collection_angle=50, convergence_angle=40)
s.metadata
```

```python
s.decomposition(normalize_poissonian_noise=True)
```

```python
s.plot_explained_variance_ratio()
```

```python
s_dn = s.get_decomposition_model(components=3)
```

```python
s_dn.plot()
```

```python
s.blind_source_separation(number_of_components=3)
```

```python
s.plot_bss_factors()
```

```python
s.plot_bss_loadings()
```

```python
s_dn.remove_background()
```

```python
Mn_peak = s_dn.isig[700.:]
```

```python
Mn_peak.plot()
```

```python
m = Mn_peak.create_model(auto_background=False)
m
```

```python
m.plot(plot_components=True)
```

```python
g1 = hs.model.components1D.GaussianHF()
m.append(g1)
m.components.GaussianHF.name = "narrow"
```

```python
m.fit_component(g1, only_current=False)
```

```python
g2 = hs.model.components1D.GaussianHF()
m.append(g2)
m.components.GaussianHF.name = "elbow"
```

```python
m.fit_component(g2, bounded=True)
```

```python
m.store(name="first_fit")
```

```python
m = m.signal.models.restore("first_fit")
```

```python
m.plot(plot_components=True)
```

```python
g3 = hs.model.components1D.GaussianHF()
m.append(g3)
m.components.GaussianHF.name = "wide"
```

```python
g1.set_parameters_not_free()
g2.set_parameters_not_free()
```
