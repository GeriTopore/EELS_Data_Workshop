# Mn oxidation state analysis

```python
s_dn.plot()
```

Here I have cropped the EELS spectrum to include only the Mn L peaks and made a ROI to investigate the EELS signal from each atomic column in one perovskite layer. You need to manually modify the ROI shape to fit each one of the layers and save it:

```python
Mn_peak = s_dn.isig[637.:655.]

roi = hs.roi.RectangularROI()
Mn_peak.plot()
s_EELS_roi = roi.interactive(Mn_peak, recompute_out_event=None)
s_EELS_roi_sum = hs.interactive(s_EELS_roi.sum, recompute_out_event=None)
s_EELS_roi_sum.plot()
```

```python
# save each layer as .hspy by moving the ROI
# s_EELS_roi_sum.save("eels_layer5.hspy")
```

```python
# loading data for each layer back in as Signal1D objects
eels_layer1 = hs.load("eels_layer1.hspy")
eels_layer2 = hs.load("eels_layer2.hspy")
eels_layer3 = hs.load("eels_layer3.hspy")
eels_layer4 = hs.load("eels_layer4.hspy")
eels_layer5 = hs.load("eels_layer5.hspy")
```

Here I start the code for making a plot to compare the spectra. I plot them in a "cascade" style and include vertical lines to highlight the positions of the peaks in the fine structure that correspond to different oxidation states:

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(7,7))
```

```python
# labels for the plot legend
labels = ['Layer 1', 'Layer 2', 'Layer 3', 'Layer 4', 'Layer 5']
```

```python
# cascade plot, padding used to specify the offset between plots (between 0 and 1)
hs.plot.plot_spectra([eels_layer1, eels_layer2, eels_layer3, eels_layer4, eels_layer5], style="cascade", fig=fig, ax=ax, padding=0.4)
```

```python
# add the vertical lines and text for annotation, set the labels, etc
ax.axvline(x=639, ls="--", c="black", lw=1)
ax.axvline(x=640.3, ls="--", c='black', lw=1)
ax.axvline(x=650.9, ls="--", c='black', lw=1)
ax.set_ylim(top=2.4)
ax.text(637.45, 2.3, "$\mathregular{Mn^{2+}}$", fontweight="bold", fontsize=14)
ax.text(640.7, 2.3, "$\mathregular{Mn^{3+}}$", fontweight="bold", fontsize=14)
ax.text(639.58, 2.2, "$\mathregular{L_3}$", fontweight="bold", color='red', fontsize=14)
ax.text(650, 2.2, "$\mathregular{L_2}$", fontweight="bold", color='red', fontsize=14)
ax.set_ylabel("Intensity (au)", fontsize=14)
ax.set_xlabel('Energy loss (eV)', fontsize=14)
ax.legend(labels, fontsize=9)
```

```python
# bbox_inches = 'tight' ensures the plot doesn't have any empty space around it when saved as an image
fig.savefig("Mn_EELS.png", dpi=500, bbox_inches='tight')
```

# Fe oxidation state analysis

```python
s_dn.plot()
```

```python
Fe_peak = s_dn.isig[702.:]
```

```python
# same thing here as with Mn
roi = hs.roi.RectangularROI()
Fe_peak.plot()
s_EELS_roi = roi.interactive(Fe_peak, recompute_out_event=None)
s_EELS_roi_sum = hs.interactive(s_EELS_roi.sum, recompute_out_event=None)
s_EELS_roi_sum.plot()
```

```python
s_EELS_roi_sum.save("Fe_eels_layer5.hspy")
```

```python
eels_layer1 = hs.load("Fe_eels_layer1.hspy")
eels_layer2 = hs.load("Fe_eels_layer2.hspy")
eels_layer3 = hs.load("Fe_eels_layer3.hspy")
eels_layer4 = hs.load("Fe_eels_layer4.hspy")
eels_layer5 = hs.load("Fe_eels_layer5.hspy")
```

Fine structure changes for Fe were more subtle, however the $L_{3}/L_{2}$ ratio clearly changes from the edge to the centre of a perovskite layer, which also suggests a change in oxidation state. The analysis and visualisation is a bit different from before. First I crop the navigation dimensions since we are only interested in a single perovskite layer. Since we are not interested in the lateral direction, I rebin the signal so the signals get summed together across the x-axis. Then, across the thickness of the film I use two 2 eV windows to calculate the intensity from the L3 and L2 peaks respectively using integration. For each pixel in the navigation dimension these intensities get divided to give us the ratio.

```python
unit = Fe_peak.inav[:, -9.2:-7.5]
```

```python
unit.plot()
```

```python
r = unit.rebin([1, 44, 258])
r = r.squeeze() # squeezing essentially removes all axes that have a size of 1, since you can't navigate across them and hence are redundant (see shape of r below)
r
```

```python
roi = hs.roi.SpanROI(left=707, right=709)
r.plot()
roi.add_widget(r, axes=["Energy loss"])
```

```python
# integrate across window
s_L3 = r.isig[roi].integrate1D(axis="Energy loss")
s_L3
```

```python
# s_L3.T.plot()
```

```python
# same thing for L2
roi = hs.roi.SpanROI(left=719.65, right=721.65)
r.plot()
roi.add_widget(r, axes=["Energy loss"])
```

```python
s_L2 = r.isig[roi].integrate1D(axis="Energy loss")
s_L2
```

```python
# s_L2.T.plot()
```

```python
# L2/L3 ratio, doing the transpose is probably redundant, but it feels weird not to
ratio = s_L3.T / s_L2.T
ratio
```

```python
# ratio.plot()
```

While I can't really see any peak shape changes with the naked eye, maybe other people can, so for the plot I include all the EELS spectra for each atomic column like before, but plot them overlapped instead. Then I include the L3/L2 intensity ratio across the columns as an inset.

```python
# import the tools we need to make the inset
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
fig, ax = plt.subplots(figsize=(7,7)) 
```

```python
# inset axis created here, loc = 1 means top right corner of the parent axis (ax)
ax_ins = inset_axes(ax, width="50%", height="40%", loc=1)
```

```python
# plot the EELS spectra, overlapped, add axis labels and legend
labels = ['Layer 1', 'Layer 2', 'Layer 3', 'Layer 4', 'Layer 5']
hs.plot.plot_spectra([eels_layer1, eels_layer2, eels_layer3, eels_layer4, eels_layer5], fig=fig, ax=ax, legend=labels)
ax.legend(labels, loc="upper left", fontsize=9)
ax.set_ylabel("Intensity (au)", fontsize=14)
ax.set_xlabel("Energy loss (eV)", fontsize=14)
ax.set_yticks([])
```

```python
# annotations for the inset
ax_ins.set_ylabel('$\mathregular{L_3 / L_2}$ intensity ratio')
ax_ins.set_xlabel('Distance (nm)')
ax_ins.plot(ratio.axes_manager[0].axis, ratio, "ro--")
ax_ins.text(-9.25, 1.305, "Layer 1", fontsize=9)
ax_ins.text(-8.88, 1.31, "Layer 2", fontsize=9)
ax_ins.text(-8.55, 1.362, "Layer 3", fontsize=9)
ax_ins.text(-8.2, 1.31, "Layer 4", fontsize=9)
ax_ins.text(-7.86, 1.3, "Layer 5", fontsize=9)
ax_ins.set_ylim(ymax=1.38)
```

```python
fig.savefig("Fe_peaks.png", dpi=500, bbox_inches='tight')
```
